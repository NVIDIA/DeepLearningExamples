```python
# read all images into np array

from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(POSTER_PATH) if isfile(join(POSTER_PATH, f))]
```

```python
len(onlyfiles)
```

```python
# read all file in dict
from PIL import Image
from tqdm import tqdm

images = {}

for f in tqdm(onlyfiles):    
    imdbID = f.split('.')[0]
    try:
        images[imdbID] = Image.open(join(POSTER_PATH, f))
    except:
        # cannot open image: take a default
        images[imdbID] = Image.open(join(POSTER_PATH, '1856057.jpg'))
```
